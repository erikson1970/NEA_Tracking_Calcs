```python
import numpy as np
import matplotlib.pyplot as plt

#write a function which creates a plot of N hexagonal close packed circles of radius r
r=1.0
N=5
def hexagon(r,N):
    #function which creates a mesh of hexagons with radius r and N hexagons
    #returns a list of x,y coordinates
    xx=[]
    yy=[]
    for ii in range(N):
        for jj in range(N):
            xx.append((ii+jj%2/2.0)*r)
            yy.append(jj*np.sqrt(3)/2.0*r)
    return xx,yy

keepers=[1,5,6,7,8,10,11,12,13,15,16,17]

xx,yy=hexagon(r,N)
xx,yy=[x for  x in zip(*[(xx[ii],yy[ii]) for ii in keepers])]
fig,ax=plt.subplots(1,1,figsize=(10,10))
ax.scatter(xx-(np.ones(len(xx))*2.0),yy-np.mean(yy))
ax.set_aspect('equal')

```
